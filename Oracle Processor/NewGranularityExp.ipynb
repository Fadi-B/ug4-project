```python

def plot_gran_results(data, true_cap, trace):
    
    #plt.figure(figsize=(10, 6))

    #up to and including 90
    x_axis = [10*x for x in np.arange(10)]
    
    #For the 1ms case
    x_axis[0] = 1

    x_axis = np.concatenate((np.array(x_axis), [100, 120, 180, 240, 300, 360, 420, 480]))
    
    thr = data[::2]
    dly = data[1::2]
    
    print(thr.shape)
    print(dly.shape)

    ### For the delay ###
    y_axis = np.mean(dly, axis=1)
    
    std_dev = np.std(dly, axis=1)

    fig, axs = plt.subplots(1, 2, figsize=(18, 6))

    axs[0].grid()

    axs[0].scatter(x_axis, y_axis, marker="x", label="Experimental")
    axs[0].errorbar(x_axis, y_axis, yerr=std_dev, fmt="o", color="b", markersize=1, capsize=3)

    axs[0].scatter(x_axis, x_axis, marker="x", color="orange", label="Ideal")

    axs[0].set_title("Self-inflicted delay vs sending granularity", fontsize=20, pad=20)
    axs[0].set_xlabel("Granularity (ms)", fontsize=16)
    axs[0].set_ylabel("Delay (ms)", fontsize=16)

    axs[0].legend(loc='lower right')

    ### For the throughput ###

    y_axis = np.mean(thr, axis=1)
    std_dev = np.std(thr, axis=1)
    
    axs[1].grid()
    axs[1].set_ylim(0,true_cap+0.05)

    axs[1].scatter(x_axis, y_axis, marker="x", label="Experimental")
    axs[1].scatter(x_axis, np.repeat(true_cap, len(x_axis)), marker="x", color="orange", label="Ideal")
    axs[1].errorbar(x_axis, y_axis, yerr=std_dev, fmt="o", color="b", markersize=1, capsize=3)

    axs[1].set_title("Average throughput vs sending granularity", fontsize=20, pad=20)
    axs[1].set_xlabel("Granularity (ms)", fontsize=16)
    axs[1].set_ylabel("Throughput (Mbit/s)", fontsize=16)

    axs[1].legend(loc='lower right')
    
    fig.savefig(trace + '-gran' + '.pdf', dpi=500)
```

```python
plot_gran_results(data, true_cap, 'TMobile-LTE')
```
